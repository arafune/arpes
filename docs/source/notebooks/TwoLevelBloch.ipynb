```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

from pulselaser.bloch import bloch, gaussian_envelope
```

```python
t_span = (-300, 2000)
rho=(0+0j, 0+0j)
t_eval=np.linspace(*t_span, 40000)
```

```python
solution = solve_ivp(bloch, t_span, rho, args=(150, 20, 0, 1e-2), vectorized=True, t_eval=t_eval)
maxy=np.max(solution.y[0])
plt.plot(solution.t, solution.y[0]/maxy, label=r"$\rho_{22}$")
plt.plot(solution.t, gaussian_envelope(solution.t, fwhm=150, t0=0), label="pulse")
plt.legend()
```

```python
solution = solve_ivp(bloch, t_span, rho, args=(150, 20, 0.3, 1e-2), vectorized=True, t_eval=t_eval)
maxy=np.max(solution.y[0])

plt.plot(solution.t-300, solution.y[0]/maxy, label=r"$\rho_{22}$")
plt.plot(solution.t-300, gaussian_envelope(solution.t, fwhm=150, t0=300), label="pulse")
plt.legend()
```
